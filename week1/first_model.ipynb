```python
from ollama import chat
```

```python
def mychat(message_log:list, new_message:dict):
    message_log.append(new_message)
    response = chat(
        model='llama3.2:latest',
        messages=message_log,
        think=False,   # disables thinking output (for now)
    )

    answer = response.message.content
    return message_log, answer


def user_msg(msg:str):
    return {"role": "user", "content":msg}

def assistant_msg(msg:str):
    return {"role": "assistant", "content":msg}

def system_msg(msg:str):
    return {"role": "system", "content":msg}
```

```python
log, answer = mychat([], user_msg("What is 2+2?"))
print(answer)
print(log)
```

```python
log.append(assistant_msg(answer))
#print(log)
log, answer = mychat(log, user_msg("Can you explain this to me?"))
print(answer)
print(log)
```
